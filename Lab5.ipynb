```python
#Импорт библиотек(все уже есть в анаконде) 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
%matplotlib inline
sns.set(color_codes=True)
```

```python
#Задание 1 
#Чтение csv-файла в датафрейм(таблица)
df=pd.read_csv('csv/methane.csv')
df.baseYear=df.baseYear.astype(str)
#10 стран с минимальными/максимальными выбросами
emismax=df.sort_values('emissions',ascending=False).loc[:,['country','emissions']].groupby('country').mean().sort_values('emissions',ascending=False).head(10)
emismin=df.sort_values('emissions',ascending=True).loc[:,['country','emissions']].groupby('country').mean().sort_values('emissions',ascending=True).head(10)
plt.pie(emismax.emissions, labels = emismax.index.values.tolist())
plt.title("10 стран с максимальными выбросами")
plt.show() 
plt.pie(emismin.emissions, labels = emismin.index.values.tolist())
plt.title("10 стран с минимальными выбросами")
plt.show() 
```

```python
#Континетны с максимальными выбросами
emismax=df.sort_values('emissions',ascending=False).loc[:,['region','emissions']].groupby('region').mean().sort_values('emissions',ascending=False)
emismin=df.sort_values('emissions',ascending=True).loc[:,['region','emissions']].groupby('region').mean().sort_values('emissions',ascending=True)
plt.pie(emismax.emissions, labels = emismax.index.values.tolist())
plt.title("Континенты с максимальными выбросами")
plt.show() 
plt.pie(emismin.emissions, labels = emismin.index.values.tolist())
plt.title("Континенты с минимальными выбросами")
plt.show() 
```
