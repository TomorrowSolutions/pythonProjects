```python
#Task9
in_a=input("a >>>")
in_b=input("b >>>")
if in_a.isdigit() and in_b.isdigit():
    a=int(in_a)
    b=int(in_b)
    sum=0;
    for n in range(a,b):
        if n%4==0:
            sum+=n
    print("Сумма ={}".format(sum))
else:
    print("Введено не число/не целое число/не положительное число")
```
